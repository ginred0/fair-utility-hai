```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

before_calibration_list = np.load("before_calibration_results.pkl",allow_pickle=True)
before_calibration_stacked_data = np.stack([df.to_numpy() for df in before_calibration_list])
before_calibration_mean_array = np.mean(before_calibration_stacked_data, axis=0)
before_calibration_std_array = np.std(before_calibration_stacked_data, axis=0)
before_calibration_mean_df = pd.DataFrame(before_calibration_mean_array, columns=before_calibration_list[0].columns)
before_calibration_std_df = pd.DataFrame(before_calibration_std_array, columns=before_calibration_list[0].columns)
```

```python

after_calibration_list = np.load("after_calibration_results.pkl",allow_pickle=True)
after_calibration_stacked_data = np.stack([df.to_numpy() for df in after_calibration_list])
after_calibration_mean_array = np.mean(after_calibration_stacked_data, axis=0)
after_calibration_std_array = np.std(after_calibration_stacked_data, axis=0)
after_calibration_mean_df = pd.DataFrame(after_calibration_mean_array, columns=after_calibration_list[0].columns)
after_calibration_std_df = pd.DataFrame(after_calibration_std_array, columns=after_calibration_list[0].columns)
```

```python

after_fair_calibration_list = np.load("after_fair_calibration_results.pkl",allow_pickle=True)
after_fair_calibration_stacked_data = np.stack([df.to_numpy() for df in after_fair_calibration_list])
after_fair_calibration_mean_array = np.mean(after_fair_calibration_stacked_data, axis=0)
after_fair_calibration_std_array = np.std(after_fair_calibration_stacked_data, axis=0)
after_fair_calibration_mean_df = pd.DataFrame(after_fair_calibration_mean_array, columns=after_fair_calibration_list[0].columns)
after_fair_calibration_std_df = pd.DataFrame(after_fair_calibration_std_array, columns=after_fair_calibration_list[0].columns)
```

```python
from matplotlib.patches import Patch 
sns.set_theme(style="whitegrid")

np.random.seed(10)

fig, axes = plt.subplots(1, 2, figsize=(10, 3))

palette1 = sns.color_palette("coolwarm", 4)
palette2 = sns.color_palette("Spectral", 4)
t = 0
custom_palette = {"Multicalibration": palette2[0], "Group-level Multicalibration": palette2[3]}  

for i in [6, 9]:
    after_calibration_data = [after_calibration_stacked_data[:, j, i] for j in range(4)]
    after_fair_calibration_data = [after_fair_calibration_stacked_data[:, j, i] for j in range(4)]

    after_calibration_flat = np.concatenate(after_calibration_data)
    after_fair_calibration_flat = np.concatenate(after_fair_calibration_data)

    category_labels_after = np.concatenate([[j+1]*len(after_calibration_stacked_data) for j in range(4)])
    category_labels_fair = np.concatenate([[j+1]*len(after_fair_calibration_stacked_data) for j in range(4)])
    df = pd.DataFrame({
        'Data': np.concatenate([after_calibration_flat, after_fair_calibration_flat]),
        'Calibration Method': ['Multicalibration'] * len(after_calibration_flat) + ['Group-level Multicalibration'] * len(after_fair_calibration_flat),
        'Experiment': np.concatenate([category_labels_after, category_labels_fair])   
    })

    sns.violinplot(x='Experiment', y='Data', bw_adjust=.5,hue='Calibration Method', data=df, ax=axes[t], linewidth=1,palette=custom_palette)

    if t == 0:
        axes[t].set_ylabel("Accuracy")
    else:
        axes[t].set_ylabel("Accuracy Disparity")
        
    t += 1

plt.tight_layout()
plt.show()
plt.savefig("results/acc_bias_cmp.pdf")
```

```python
sns.set_theme(style="whitegrid")

np.random.seed(10)

fig, axes = plt.subplots(1, 2, figsize=(10, 3))
palette1 = sns.color_palette("coolwarm", 4)
palette2 = sns.color_palette("Spectral", 4)
custom_palette = {
    "No-calibration": "black",
    "Multicalibration": palette2[0],
    "Group-level Multicalibration": palette2[3]
}  

t = 0

axes[1].grid(True, zorder=-2)  

for i in [6, 9]:
    before_calibration_data = [ np.mean(before_calibration_stacked_data[:, j, i],axis=0)  for j in range(4)]
    after_calibration_data = [after_calibration_stacked_data[:, j, i] for j in range(4)]
    after_fair_calibration_data = [after_fair_calibration_stacked_data[:, j, i] for j in range(4)]
    
    num_samples = after_calibration_stacked_data.shape[0]
    before_calibration_data_repeated = [
        np.full(num_samples, before_calibration_data[j]) for j in range(4)
    ]
    
    calibration_methods = ['No-calibration', 'Multicalibration', 'Group-level Multicalibration']
    data_list = []
    for method, data_array in zip(
        calibration_methods,
        [before_calibration_data_repeated, after_calibration_data, after_fair_calibration_data]
    ):
        for j in range(4): 
            data = data_array[j]
            df_temp = pd.DataFrame({
                'Data': data,
                'Calibration Method': method,
                'Experiment': j + 1 
            })
            data_list.append(df_temp)
    df = pd.concat(data_list, ignore_index=True)
    
    
    sns.pointplot(
        x='Experiment',
        y='Data',
        hue='Calibration Method',
        data=df[df['Calibration Method'] == 'No-calibration'],
        ax=axes[t],
        palette=custom_palette,
        markers='^',
        scale=1,
        linestyles='',
        dodge=0.2
    )

    sns.boxplot(
        x='Experiment',
        y='Data',
        hue='Calibration Method',
        data=df[df['Calibration Method'] != 'No-calibration'],
        ax=axes[t],
        linewidth=1,
        palette=custom_palette,
        showfliers=False,
        zorder=3
    )
    

    
    if t == 0:
        axes[t].set_ylabel("Utility of $T=\pi(h,a)$",fontsize=12, fontweight='bold')
        axes[t].legend([], [], frameon=False) 
    else:
        axes[t].axhline(y=0, color='red', linewidth=1, linestyle='--', zorder=1) 
        axes[t].set_ylabel("Utility Disparity of $T=\pi(h,a)$",fontsize=12, fontweight='bold')
        
    axes[t].set_xlabel("Experiment", fontsize=12, fontweight='bold')    
    
    t += 1
    
axes[1].set_ylim([-0.04,0.04])

plt.legend(  loc='best',fontsize=10)    
plt.tight_layout()
plt.savefig("results/acc_bias_cmp.pdf")
plt.show()
```

```python
from matplotlib.patches import Patch 
sns.set_theme(style="whitegrid")

np.random.seed(10)
fig, axes = plt.subplots(1, 2, figsize=(10, 3))

palette1 = sns.color_palette("coolwarm", 4)
palette2 = sns.color_palette("Spectral", 4)

hatch_patterns = {"No-calibration": None, "Multicalibration": '\\', "Group-level Multicalibration": '/'}

t = 0
custom_palette = {"No-calibration": palette1[0],"Multicalibration": palette2[0], "Group-level Multicalibration": palette2[3]} 
for i in [6, 9]:
    before_calibration_data = [[np.mean(before_calibration_stacked_data[:, j, i],axis=0)] for j in range(4)]
    after_calibration_data = [[np.mean(after_calibration_stacked_data[:, j, i],axis=0)] for j in range(4)]
    after_fair_calibration_data = [[np.mean(after_fair_calibration_stacked_data[:, j, i],axis=0)] for j in range(4)]

    before_calibration_flat = np.concatenate(before_calibration_data)
    after_calibration_flat = np.concatenate(after_calibration_data)
    after_fair_calibration_flat = np.concatenate(after_fair_calibration_data)

    category_labels_before = np.concatenate([[j+1] for j in range(4)])
    category_labels_after = np.concatenate([[j+1] for j in range(4)])
    category_labels_fair = np.concatenate([[j+1] for j in range(4)])
    df = pd.DataFrame({
        'Data': np.concatenate([before_calibration_flat,after_calibration_flat, after_fair_calibration_flat]),
        'Calibration Method': ['No-calibration'] * len(before_calibration_flat)+['Multicalibration'] * len(after_calibration_flat) + ['Group-level Multicalibration'] * len(after_fair_calibration_flat),
        'Experiment': np.concatenate([category_labels_before,category_labels_after, category_labels_fair])  
    })

    sns.barplot(x='Experiment', y='Data', hue='Calibration Method', data=df, ax=axes[t],palette=custom_palette)

    for patch, label in zip(axes[t].patches, df['Calibration Method']):
        patch.set_hatch(hatch_patterns[label])
        
    if t == 0:
        axes[t].set_ylabel("Final Decision Acc.")
        axes[t].legend([], [], frameon=False) 
    else:
        axes[t].set_ylabel("Final Decision  Acc. Disp.")
        axes[t].axhline(y=0, color='red', linewidth=1, linestyle='--')
        
    t += 1

legend_patches = [Patch(facecolor=custom_palette[method], hatch=hatch_patterns[method], label=method) for method in custom_palette]

plt.legend(handles=legend_patches,  loc='best')    

plt.tight_layout()
plt.savefig("results/ai_bias_cmp.pdf")
plt.savefig("results/ai_bias_cmp.svg")
plt.show()
```

```python
from matplotlib.patches import Patch 
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)

np.random.seed(10)
fig, axes = plt.subplots(1, 2, figsize=(7, 3))

palette1 = sns.color_palette("coolwarm", 4)
palette2 = sns.color_palette("Spectral", 4)

hatch_patterns = {"Human": None}

t = 0
custom_palette = {"Human": '#ebd9a5' } 
for i in [5, 8]:
    before_calibration_data = [[before_calibration_stacked_data[0, j, i]] for j in range(4)]
    
    before_calibration_flat = np.concatenate(before_calibration_data)

    category_labels_before = np.concatenate([[j+1] for j in range(4)])

    df = pd.DataFrame({
        'Data': np.concatenate([before_calibration_flat]),
        'Calibration Method': ['Human'] * len(before_calibration_flat),
        'Experiment': np.concatenate([category_labels_before])  
    })

    sns.barplot(x='Experiment', y='Data', hue='Calibration Method', data=df, ax=axes[t],palette=custom_palette,width=0.6)

    hue_order = df['Calibration Method'].unique()

    edge_colors = [custom_palette[hue] for hue in hue_order]

    for c, container in enumerate(axes[t].containers):
        for bar in container.patches:
            bar.set_edgecolor(edge_colors[c])  
            bar.set_linewidth(1)               

    for patch, label in zip(axes[t].patches, df['Calibration Method']):
        patch.set_hatch(hatch_patterns[label])
        
    if t == 0:
        axes[t].set_ylabel("Utility of $T=\pi(h)$", fontsize=12, fontweight='bold')  
        
    else:
        axes[t].set_ylabel("Utility Disparity of $T=\pi(h)$", fontsize=12, fontweight='bold')  
        axes[t].axhline(y=0, color='red', linewidth=1, linestyle='--',zorder=0)
        
    axes[t].set_xlabel("Experiment", fontsize=12, fontweight='bold')  
    axes[t].legend([], [], frameon=False) 

        
    t += 1

legend_patches = [Patch(facecolor=custom_palette[method], hatch=hatch_patterns[method], label=method) for method in custom_palette]


plt.tight_layout()
plt.savefig("results/human_performance_cmp.pdf")
plt.savefig("results/human_performance_cmp.svg")
plt.show()
```
